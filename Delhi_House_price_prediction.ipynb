```python
 #preprocessing the data
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/Delhi.csv')
print(data.head())
data.shape
```

```python
data.info()
```

```python
#print(data.head())
security=data['24X7Security']
lift=data['LiftAvailable']
resale=data['Resale']
car_parking=data['CarParking']
MaintenanceStaff =data['MaintenanceStaff']
VaastuCompliant=data['VaastuCompliant']
data1=data.iloc[:,:4]
data1['security']=security
data1['lift_available']=lift
data1['Resale']=resale
data1['CarParking']=car_parking
data1['MaintenanceStaff']=MaintenanceStaff
data1['VaastuCompliant']=VaastuCompliant
print(data1)
```

```python
for c in data1.columns:
  print(data1[c].value_counts())
  print("*"*20)
```

```python
arr=[]
for i in range(2002,4998):
  arr.append(i)
data1=data1.drop(arr)
```

```python
print(data1.isna().sum())
data1
```

```python

locationcounts=data1['Location'].value_counts()
locationcount_lessthan_10=locationcounts[locationcounts<=10]
print(locationcount_lessthan_10)
```

```python
data1['Location']=data['Location'].apply(lambda x:"Other" if x in locationcount_lessthan_10 else x)
data1['Location'].value_counts()
data1['Location'].value_counts()
```

```python
data1.head()
```

```python
data1['security']=data1['security'].apply(lambda x: "Yes" if x == 1 else "No" )
data1['lift_available']=data1['lift_available'].apply(lambda x: "Yes" if x == 1 else "No" )
data1['Resale']=data1['Resale'].apply(lambda x: "Yes" if x == 1 else "No" )
data1['CarParking']=data1['CarParking'].apply(lambda x: "Yes" if x == 1 else "No" )
data1['MaintenanceStaff']=data1['MaintenanceStaff'].apply(lambda x: "Yes" if x == 1 else "No" )
data1['VaastuCompliant']=data1['VaastuCompliant'].apply(lambda x: "Yes" if x == 1 else "No" )
data1
```

```python
x=data1.iloc[:,1:]
y=data1['Price']
print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
```

```python
#linear regression
#col_trans=make_column_transformer((OneHotEncoder(sparse=False),['Location'],['security'],['lift_available'],['Resale'],['CarParking'],['MaintenanceStaff'],['VaastuCompliant']),remainder='passthrough')
col_trans = make_column_transformer(
    (OneHotEncoder(sparse=False), ['Location', 'security', 'lift_available', 'Resale', 'CarParking', 'MaintenanceStaff', 'VaastuCompliant']),
    remainder='passthrough')
sr=StandardScaler()
lr=LinearRegression()
pipe=make_pipeline(col_trans,sr,lr)
pipe.fit(x_train,y_train)
y_predict=pipe.predict(x_test)
r2_score(y_test,y_predict)
```

```python
#applying lasso
lasso=Lasso()
pipe=make_pipeline(col_trans,sr,lasso)
pipe.fit(x_train,y_train)
y_predict_las=pipe.predict(x_test)
r2_score(y_test,y_predict_las)
```

```python
#applying ridge
ridge=Ridge()
pipe=make_pipeline(col_trans,sr,ridge)
pipe.fit(x_train,y_train)
y_predict_ridge=pipe.predict(x_test)
print(x_test.head())
r2_score(y_test,y_predict_ridge)
```
